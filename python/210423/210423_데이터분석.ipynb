```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import font_manager,rc


#코로나로 인한 감정 변화에 따른 미디어콘텐츠 이용량 변화

plt.rcParams['figure.figsize']=[20,10]
font_name=font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)

data = pd.read_csv("한국언론진흥재단_코로나19이후국민의일상변화조사_코로나19로 인한 감정 변화_20201231.csv",encoding='cp949')
print(data)

emo_list = list(data['항목'])
inc_emo_list = list(data['증가'])
sam_emo_list = list(data['변화 없음'])
dec_emo_list = list(data['감소'])

#print(inc_emo_list)
#type(inc_emo_list[0])

#x = range(len(emo_list))

a_bottom = np.add(dec_emo_list, sam_emo_list)

plt.bar(emo_list, dec_emo_list, label = '감소(%)')
plt.bar(emo_list, sam_emo_list, bottom = dec_emo_list, label = '변화 없음(%)')
plt.bar(emo_list, inc_emo_list, bottom = a_bottom, label = '증가(%)')

plt.title('코로나 19로 인한 감정 변화 추이', fontsize = 30)
plt.xlabel('감정', fontsize = 20)
plt.ylabel('감정 변화 비율', fontsize = 20)
plt.legend(bbox_to_anchor=(1.1, 1))
plt.axhline(y = 50, color='r', linewidth=1)

#print(emo_list)
```
